# Flattening

```python
flatten_size_zayn=image_size_zayn[0]*image_size_zayn[1]
flatten_size_zayn
```

```python
flatten_size_lily=image_size_lily[0]*image_size_lily[1]
flatten_size_lily
```

```python
flatten_size_niall=image_size_niall[0]*image_size_niall[1]
flatten_size_niall
```

```python
for i in range(len_of_images_zayn):
    zayn_gray[i]=np.ndarray.flatten(zayn_gray[i].reshape(flatten_size_zayn,1))
```

```python
zayn_gray=np.dstack(zayn_gray)
```

```python
zayn_gray=np.rollaxis(zayn_gray, axis=2,start=0)
zayn_gray.shape
```

```python
zayn_gray=zayn_gray.reshape(len_of_images_zayn,flatten_size_zayn)
zayn_gray.shape
```

```python
lily_gray=np.dstack(lily_gray)
```

```python
lily_gray=np.rollaxis(lily_gray, axis=2,start=0)
lily_gray.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
from sklearn import svm
```

```python
clf=svm.SVC()
clf.fit(x_train,y_train)
```

```python
y_pred=clf.predict(x_test)
```

```python
y_pred
```

```python
from sklearn import metrics
```

```python
accuracy=metrics.accuracy_score(y_test,y_pred)
```
